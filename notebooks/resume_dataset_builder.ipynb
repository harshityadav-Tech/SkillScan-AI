```python
import os 
import pandas as pd 
from docx import Document
resume_folder ="../Resumes"
files = os.listdir(resume_folder)
print("Total number of resumes: ", len(files))
print("First 5 resumes: ", files[:5])
```

```python
sample_resume = files[0]
resume_path = os.path.join(resume_folder, sample_resume)
print("Reading resume: ")
print(resume_path)
doc = Document(resume_path)
resume_text = ""
for para in doc.paragraphs:
    resume_text += para.text + "\n"
print("\nResume Text: \n")
print(resume_text[:1000])  # Print first 1000 characters of the resume text
```

```python
resume_data = []
for file in files:
    file_path = os.path.join(resume_folder, file)
    try:
        doc = Document(file_path)
        text = ""
        for para in doc.paragraphs:
            text += para.text + "\n"
        resume_data.append({"filename": file, "resume_text": text})
    except Exception as e:
        print(f"Error reading {file}: {e}")
df = pd.DataFrame(resume_data)
print("DataFrame created with resume data:")
print("Total resumes processed: ", len(df))
df
```

```python
def create_category(filename):
    filename = filename.lower()
    if "java" in filename:
        return "Java Developer" 
    elif "python" in filename:
        return "Python Developer"
    elif "analyst" in filename:
        return "Data Analyst"
    elif "developer" in filename:
        return "Software Developer"
    elif "manager" in filename:
        return "Project Manager"
    else:
        return "Other"
df["category"] = df["filename"].apply(create_category)
df[["filename", "category"]].head(20)
```

```python
import re 
def clean_resume_text(text):
    text = text.lower() 
    text = re.sub(r'\S+@\S+', ' ', text)  
    text = re.sub(r'd+', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
df["cleaned_resume_text"] = df["resume_text"].apply(clean_resume_text)
df[["filename", "cleaned_resume_text"]].head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word not in stop_words]
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text
df["processed_resume"] = df["cleaned_resume_text"].apply(preprocess_text)
df[["category","processed_resume"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(df["processed_resume"])
y = df["category"]
print("TF-IDF features created. Shape: ", X.shape)
print("Total categories:", y.nunique())
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report,accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy: ", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
print("Random Forest Model Accuracy: ", rf_accuracy)
```

```python
def predict_resume_category(resume_text):
    cleaned_text = clean_resume_text(resume_text)
    processed_text = preprocess_text(cleaned_text)
    vectorized_text = tfidf.transform([processed_text])
    prediction = rf_model.predict(vectorized_text)
    return prediction[0]
```

```python
sample_resume ="""
Python developer with 5 years of experience in building web applications using Django and Flask. Proficient in data analysis and machine learning with libraries like Pandas, NumPy, and Scikit-learn. Strong background in software development and project management.
"""
predicted_category = predict_resume_category(sample_resume) 
print("Predicted Category for Sample Resume: ", predicted_category)
```

```python
skills = ["python", "java","c++","javascript","html","css","react","nodejs","flask","django","tensordlow","pytorch","machine learning","nlp","data science","sql","mongodb","mysql","git","github","aws","docker","kubernetes","pandas","numpy","scikit-learn"]
print("Skills list created with ", len(skills), " skills.")
```

```python
def extract_skills(resume_text):
    resume_text = resume_text.lower()
    found_skills=[]
    for skill in skills:
        if skill.lower() in resume_text:
            found_skills.append(skill)
    return found_skills
```

```python
sample_resume ="""Python Developer with experience in machine learning and data analysis using libraries like Pandas and NumPy. Proficient in web development with Flask and Django. Skilled in SQL, Git, and AWS. """
skills_found = extract_skills(sample_resume)
print("Skills found in sample resume: ", skills_found)
```

```python
job_description = """We are looking for a Python Developer with experience in machine learning and data analysis. The ideal candidate should be proficient in web development using Flask and Django, and have strong skills in SQL, Git, and AWS."""
skills_in_job_description = extract_skills(job_description)
print("Skills found in job description: ", skills_in_job_description)
```

```python
resume_skills = extract_skills(sample_resume)
matched_skills = []
for skill in skills_in_job_description:
    if skill in resume_skills:
        matched_skills.append(skill)
ats_score = len(matched_skills) / len(skills_in_job_description) * 100
print("Matched Skills: ", matched_skills)
print("ATS Score: ", ats_score, "%")
```
